```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
from nsetools import Nse
nseObject = Nse()
all_stock_codes = nseObject.get_stock_codes()
```

```python
len(all_stock_codes)
```

```python
listStockCodes = list(all_stock_codes.keys())
listStockCodes.pop(0)
```

```python
nse = {new_list: [] for new_list in listStockCodes}
for i in listStockCodes:
    try:
        ticker = yf.Ticker(i + '.NS')
        tickerHistory = ticker.history(interval="1d", start="2022-06-22", end="2022-06-25")
        Open = tickerHistory.iloc[-2]['Open']
        Close = tickerHistory.iloc[-1]['Close']
        priceGrowth = (Close - Open)/Open*100
        TradeVolume1 = tickerHistory.iloc[-2]['Volume']
        TradeVolume2 = tickerHistory.iloc[-1]['Volume']
        volumeGrowth = (TradeVolume2 - TradeVolume1)/TradeVolume1 * 100
        nse[i] = [Open,Close,priceGrowth, TradeVolume1, TradeVolume2, volumeGrowth]
    except IndexError:
        nse.pop(i)
        continue
```

```python
len(nse)
```

```python
nse
```

```python
stocks = pd.DataFrame.from_dict(nse)
```

```python
stocks.head()
```

```python
stocks2 = stocks.transpose()
```

```python
stocks2.rename(columns = {0:'Open', 1:'Close',2:'Price Growth', 3:'Trade Volume Day 1', 4: 'Trade Volume Day 2', 5: 'Volume Growth Percent'}, inplace = True)
```
